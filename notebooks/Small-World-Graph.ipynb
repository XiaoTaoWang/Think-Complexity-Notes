## Wattz-Strogatz Graph

A typical small-world graph has two properties:
- First, nodes in the graph tend to be clustered. In other words, small world graphs tend to contain sub-graphs, in which nodes are all connected to each other. Suppose your friends are also friends of each other.
- Second, pairs of nodes are always connected by at least one short path, that is, the mean shortest-path length between pairs of nodes is much shorter than expected. This property is somewhat surprising because we know we always live near our friends, and in a graph with only local connections, the path lengths tend to increase in proportion to geographical distance, right?

To quantify these two properties, Watts and Strogatz defined two measures: clustering coefficient *C* and path length *L*:
- Clustering coefficient quantifies the likehood that two nodes conected to the same node are also connected each other. Specifically, suppose a node *v* has *k* neighbors, there're at most $\frac{k(k-1)}{2}$ edges between the neighbors, the fraction of edges actually existing is defined as the local clustering coefficient for *v*. Then the clustering coefficient of a graph is defined as the average for all local clustering coefficients.
- Path length is defined as the average distance between any two nodes.

They first considered two traditional graph model: random graph and regular graph. In a random graph, which I've included in previous notes, nodes are connected at random. In a regular graph, every node has the same number of neighbors. The two kind of graphs show opposite properties: random graph has low path lengths and low clustering, while regular graph high clustering and high path lengths.

Maybe because small-world graph's properties fall in the middle between regular graph and random graph, Watts and Strogatz combined these two models and proposed a generative model for small-world graphs:
- Start with a regular graph.
- Iterate edges and "rewire" a edge with probability *p* by replacing it with random edges.

*p* controls how random the graph is, with *p* = 0, the graph is regular, with *p* = 1, the graph is random.

Watts and Strogatz showed that with small *p*, the yield graphs have high clustering like a regular graph, and low path lengths like a random graph.

## Small-world graph in *networkx*

*networkx* provides convenient functions for investigating properties of small-world graphs.

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 10
k = 4
ps = [0, 0.2, 1]
fig = plt.figure(figsize=(10, 3))
for i in range(3):
    ax = plt.subplot(1, 3, i+1)
    G = nx.watts_strogatz_graph(n, k, ps[i])
    ax.set_title('p = %.1f' % ps[i])
    nx.draw_circular(G, node_size=40)
plt.show()
```

The figure above shows Wattz-Strogatz graphs with $n = 10$, $k = 4$, and a range of $p$. When $p = 0$, the graph is a kind of regular graph(ring lattice), when $p = 1$, it's completely random. 

```python
import numpy as np

def possible_pair(nodes):
    for i, v in enumerate(nodes):
        for j, w in enumerate(nodes):
            if i > j:
                yield v, w


def average_path_length(G):
    pool = nx.shortest_path_length(G)
    lengths = [pool[v][w] for v,w in possible_pair(G)]
    return np.mean(lengths)
```

```python
n = 1000
k = 10
ps = [0, 0.02, 1]
print '\t'.join(['p', 'Clustering Coefficient', 'Path Length'])
print '-'*43
for p in ps:
    G = nx.watts_strogatz_graph(n, k, p)
    C = nx.average_clustering(G)
    L = average_path_length(G)
    print '%s\t\t%.3f\t\t   %.3f' % (p, C, L)
```

*average_clustering* function returns clustering coefficient directly. And *shortest_path_length* returns a dictionary of dictionaries containing shortest path lengths for all pairs.

```python
n_iter = 3
ps = np.logspace(-4, 0, 10)

Cp = np.zeros((n_iter, ps.size))
Lp = np.zeros((n_iter, ps.size))

for i in range(n_iter):
    for j in range(ps.size):
        G = nx.watts_strogatz_graph(n, k, ps[j])
        Cp[i,j] = nx.average_clustering(G)
        Lp[i,j] = average_path_length(G)

# Normalized by L0, C0
for i in range(1, Cp.shape[1]):
    Cp[:,i] = Cp[:,i] / Cp[:,0]
    Lp[:,i] = Lp[:,i] / Lp[:,0]
Cp[:,0] = 1; Lp[:,0] = 1

# Averages over all iterations
Cs = np.mean(Cp, axis=0)
Ls = np.mean(Lp, axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(ps, Cs, marker='s', color='#FED9A6', markersize=7)
plt.plot(ps, Ls, marker='o', color='#DECBE4', markersize=7)
plt.xscale('log')
plt.text(0.01, 0.8, 'C(p) / C(0)', fontsize=12)
plt.text(0.001, 0.22, 'L(p) / L(0)', fontsize=12)
plt.xlabel('p', fontsize=16)
plt.show()
```

As shown in above, as *p* increases, the path lengths drops quickly, while clustering coefficient decreases much more slowly, as a result, there a wide range of *p* where Wattz-Strogatz graphs show the properties of small-world graphs. Think it more deeply, small-world graphs in Wattz-Strogatz model are caused by the introduction of a few long-range edges, and each such edge has a highly nonlinear effect on *L*, because it affect not just distance between nodes it connects, but also their neighbors, neighbors of neighbors, and so on. On the other hand, removing an edge in the regular graph has at most linear effect on *C*, hence $C(p)$ remains generally unchanged for small *p* while $L(p)$ drops rapidly.
