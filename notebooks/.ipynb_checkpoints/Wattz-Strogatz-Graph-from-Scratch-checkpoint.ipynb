## Ring Lattice

Wattz and Strogatz use a kind of regular graph - *ring lattice* as the basis of their model. In a ring lattice, nodes are arranged in a circle with each node connecting the k nearest neighbors. Here, I define a function named *ring_lattice_edges* which takes a list of nodes and an interger number *k*, and yields one edge of corresponding ring lattice at a time.

```python
def ring_lattice_edges(nodes, k):
    half = k // 2
    n = len(nodes)
    for i, v in enumerate(nodes):
        for j in range(i+1, i+half+1):
            w = nodes[j % n]
            yield v, w
```

Then given *n* and *k*, we can generate a ring lattice in a way that we are familar with:

```python
import networkx as nx

def make_ring_lattice(n, k):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    edges = ring_lattice_edges(nodes, k)
    G.add_edges_from(edges)
    return G
```

```python
import matplotlib.pyplot as plt

G = make_ring_lattice(10, 4)
nx.draw_circular(G, node_size=100)
plt.show()
```
